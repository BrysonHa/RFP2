```python
import pandas as pd

# Load the CSV file
file_name = "trackslater1960.csv"
df = pd.read_csv(file_name)

# Take 10,000 random songs (or less if the file has fewer rows)
sample_size = 10000
random_songs = df.sample(n=min(sample_size, len(df)), random_state=42)

# Print ALL characteristics of the selected songs
pd.set_option("display.max_columns", None)  # show all columns
pd.set_option("display.width", None)        # avoid cutting off lines
pd.set_option("display.max_rows", None)     # show all rows (WARNING: huge output)

print(random_songs)

# OPTIONAL: Save the random songs to a new CSV file
random_songs.to_csv("random_10000_songs.csv", index=False)
print("\nSaved as: random_10000_songs.csv")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load files
df1 = pd.read_csv("Hey_Oh_River.csv")          # has "Tempo"
df2 = pd.read_csv("random_10000_songs.csv")    # has "tempo"

# Rename Tempo -> tempo so both match
df1 = df1.rename(columns={"Tempo": "tempo"})

# Convert to numeric
df1["tempo"] = pd.to_numeric(df1["tempo"], errors="coerce")
df2["tempo"] = pd.to_numeric(df2["tempo"], errors="coerce")

# Drop missing values
tempo1 = df1["tempo"].dropna()
tempo2 = df2["tempo"].dropna()

# Box plot comparison
plt.figure(figsize=(8, 5))
plt.boxplot([tempo1, tempo2], labels=["Hey_Oh_River", "Random 10000 Songs"])
plt.title("Tempo Comparison (Box Plot)")
plt.ylabel("Tempo (BPM)")
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import re

# -------------------------------
# 1) Function to clean column names
# -------------------------------
def clean_column_name(col):
    col = col.strip().lower()                 # lowercase + remove outside spaces
    col = col.replace(" ", "_")               # spaces -> _
    col = re.sub(r"[()]", "", col)            # remove parentheses
    col = re.sub(r"[^a-z0-9_]", "", col)      # remove other symbols
    col = re.sub(r"_+", "_", col)             # remove double __
    return col


# -------------------------------
# 2) Load files
# -------------------------------
df_hey = pd.read_csv("Hey_Oh_River.csv")
df_rand = pd.read_csv("random_10000_songs.csv")


# -------------------------------
# 3) Clean column names in BOTH dataframes
# -------------------------------
df_hey.columns = [clean_column_name(c) for c in df_hey.columns]
df_rand.columns = [clean_column_name(c) for c in df_rand.columns]


# -------------------------------
# 4) Features you want to compare (in standardized format)
#    (after cleaning they should look like this)
# -------------------------------
features = [
    "duration_ms",
    "popularity",
    "danceability",
    "energy",
    "key",
    "loudness",
    "mode",
    "speechiness",
    "acousticness",
    "instrumentalness",
    "liveness",
    "valence"
]

# -------------------------------
# 5) Check which features exist in BOTH files
# -------------------------------
common_features = [f for f in features if f in df_hey.columns and f in df_rand.columns]

print("✅ Features found in BOTH files:")
print(common_features)

missing_hey = [f for f in features if f not in df_hey.columns]
missing_rand = [f for f in features if f not in df_rand.columns]

print("\n❌ Missing from Hey_Oh_River.csv:", missing_hey)
print("❌ Missing from random_10000_songs.csv:", missing_rand)


# -------------------------------
# 6) Make boxplots for each feature
# -------------------------------
for feature in common_features:
    # convert to numeric
    hey_vals = pd.to_numeric(df_hey[feature], errors="coerce").dropna()
    rand_vals = pd.to_numeric(df_rand[feature], errors="coerce").dropna()

    plt.figure(figsize=(7, 4))
    plt.boxplot([hey_vals, rand_vals], labels=["Hey_Oh_River", "Random 10000 Songs"])
    plt.title(f"Comparison of {feature} (Box Plot)")
    plt.ylabel(feature)
    plt.grid(True)
    plt.show()
```
